# Machine learning linearRegression- Dwelling and industries

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
```

```python
data = pd.read_csv('machinedata.csv')
```

```python
data.head(3)
```

```python
# Encoding 'clue_small_area'
le = preprocessing.LabelEncoder()
data['clue_small_area'] = le.fit_transform(data['clue_small_area'].values)
```

```python
data.head(3)
```

```python
data.describe()
```

```python
# All the values are going to be scaled except for target value
scaler = StandardScaler()
data.iloc[:,2:-2] = scaler.fit_transform(data.iloc[:,2:-2])
```

```python
data.describe()
```

```python
X_train, X_test, y_train, y_test = train_test_split(data.iloc[:,:-2], data.iloc[:,-1], test_size=0.33, random_state=42)
```

```python
# creating model
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
score = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {score}")
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```
